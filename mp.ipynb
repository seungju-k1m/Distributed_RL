```python
import torch
import torch.multiprocessing as mp
# import multiprocessing as mp
```

```python
import os
import time
from _pickle import loads
```

```python
def preprocessBatch(path):
    sample = []
    with open(path, "rb") as f:
        x = f.read()
        x = loads(x)
        x = torch.tensor(x).float().cpu()
        x.share_memory_()
        sample.append(x)
        f.close()
    return sample
```

```python
if __name__ == "__main__":
    q = mp.Queue()
    x = time.time()
    paths = os.listdir('./TestData/Image')
    f = lambda x : './TestData/Image/' + x
    paths = list(map(f, paths))
    pools = mp.Pool(processes=4)
    outputs = pools.map(preprocessBatch, paths)

    print(time.time() - x)
```
